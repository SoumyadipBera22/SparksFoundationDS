## Exploratory Data Analysis - Retail

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('SampleSuperstore.csv')
```

```python
df.head()
```

```python
df.info()
```

## Missing Values Check

```python
sns.heatmap(df.isna(),yticklabels=False)
```

```python
df.isna().sum()
```

## Univariate Exploratory Data Analysis

```python
sns.countplot('Ship Mode',data=df)
```

```python
# Max number of buyers opt for "Standard Class"
```

```python
sns.countplot('Segment',data=df)
```

```python
# The segment with the maximum count is "Consumer"
```

```python
sns.countplot('Region',data=df)
```

```python
# Maximum number of consumers are from the West region, followed by East
```

```python
plt.figure(figsize=(15,12))
sns.countplot('State',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
# The greatest numbers of consumers are from California and New York
```

```python
df['Category'].value_counts().plot(kind='pie', autopct='%.2f')
```

```python
# The category most dealt with is Office Supplies
```

```python
plt.figure(figsize=(15,12))
df['Sub-Category'].value_counts().plot(kind='pie', autopct='%.2f')
plt.show()
```

```python
# The sub-categories most dealt with are Paper, Furnishings, Phones, Storage, Art and Accessories
```

## Bi-variate EDA

```python
sns.barplot(x='Ship Mode',y='Sales',data=df)
```

```python
sns.barplot(x='Ship Mode',y='Discount',data=df)
```

```python
sns.barplot(x='Ship Mode',y='Profit',data=df)
```

```python
sns.barplot(x='Segment',y='Sales',data=df)
```

```python
sns.barplot(x='Segment',y='Profit',data=df)
```

```python
sns.barplot(x='Region',y='Sales',data=df)
```

```python
sns.barplot(x='Region',y='Profit',data=df)
```

```python
# Although more sales are made in the South, the profit made from the West exceelds that from the South
```

```python
sns.barplot(x='Category',y='Sales',data=df)
```

```python
sns.barplot(x='Category',y='Profit',data=df)
```

```python
# Interestingly, although maximum sales are made in the Furniture category, the profits made are minimum among the 3 categories
```

```python
plt.figure(figsize=(15,12))
sns.barplot(x='State',y='Profit',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
# Most profitable state - Vermont
# States facing losses - Ohio, Colorado
```
